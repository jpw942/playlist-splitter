```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = ["I love pizza", "Pizza is great", "The car is red"]
embeddings = model.encode(sentences)

for sentence, embedding in zip(sentences, embeddings):
    print(f"{sentence}: {embedding[:5]}...")  # prints first 5 numbers of each vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity

pairs = [
    ("I love pizza", "Pizza is great"),
    ("I love pizza", "The car is red"),
    ("Pizza is great", "The car is red"),
]

for a, b in pairs:
    idx_a = sentences.index(a)
    idx_b = sentences.index(b)
    score = cosine_similarity([embeddings[idx_a]], [embeddings[idx_b]])[0][0]
    print(f'"{a}" vs "{b}": {score:.4f}')
```
